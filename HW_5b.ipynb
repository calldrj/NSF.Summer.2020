```python
# Convert Chicago colums to a numpy array
import numpy as np
np_Chicago = np.array(df_3_cities['Chicago'])
```

```python
# Compute the mean and standard deviation of column Chicago using numpy
print('Chicago col has mean = {:.3f} and std = {:.3f}'.format(np_Chicago.mean(), np_Chicago.std()))
```

```python
# Double check the mean and standard deviation of column Chicago using pandas
df_3_cities.Chicago.describe()
```

```python
# compute the correlation between two columns (Chicago and Los Angeles) using np.corrcoef
np_LA = np.array(df_3_cities['Los Angeles'])
print(np.corrcoef(np_LA, np_Chicago))
```
