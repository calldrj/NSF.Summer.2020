```python
import mysql.connector
from mysql.connector import Error
import pandas as pd

# Fill in the database you want to work on
working_database = 'Xgridz'

# Fill in your mySQL root pass
my_password = '*****'


try:
    """CONNECT TO DATABASE XGRIDZ"""
    connection = mysql.connector.connect(host='localhost',
                                         database=working_database,
                                         user='root',
                                         password=my_password)

    if connection.is_connected():
        
        # mysqli_set_charset(connection, ”utf8mb4”);
        
         # Show the server infomation 
        db_info = connection.get_server_info()
        print('Connected to MySQL Server version ', db_info)
        
        # Establish the connection for future queries
        cursor = connection.cursor()
        
        # Show the database infomation 
        cursor.execute('select database();')
        record = cursor.fetchone()
        print('Connected to database: ', record)
        cursor.execute('show tables;')
        tables = cursor.fetchall()
        print('Tables in {} schema: {}\n\n'.format(working_database, tables))
        
        """INSERT DATA TO XGRIDZ"""
        key = '*****'
        # Insert student data from a csv file into student table
        student = pd.read_csv('student.csv')
        Q = ["INSERT INTO student (name, age) \
              VALUES ('{}', AES_ENCRYPT('{}', '{}'));" \
             .format(s[0], s[1], key) for s in student[:10].values]
        
        for q in Q:
            cursor.execute(q)
        
         # Insert Grid Test 1 data from a csv file into test1 table
        test1 = pd.read_csv('test1.csv')
        Q = ["INSERT INTO test1 (date, correct, total, weight) \
              VALUES ('{}', '{}', '{}', '{}');".\
              format(t[0], t[1], t[2], t[3]) for t in test1[:10].values]
    
        for q in Q:
            cursor.execute(q)
        
        
        """RUN THE QUERY OF REQUEST"""
        q = "SELECT test1.date AS test_date, s.ID AS student_ID, s.name, \
                        test1.correct AS correct_1, test1.total AS total_1, s.age \
                 FROM student AS s \
                              INNER JOIN test1 ON s.ID = test1.student_ID;"
        cursor.execute(q)
        cursor.fetchall()
        join_table = pd.read_sql_query(q, connection)
        print('Data of age has been encrypted:\n')
        print(join_table)
        
        q = "SELECT test1.date AS test_date, s.ID AS student_ID, s.name, \
                    test1.correct AS correct_1, test1.total AS total_1, \
                    AES_DECRYPT(s.age, '{}') AS decr_age\
             FROM student AS s \
                          INNER JOIN test1 ON s.ID = test1.student_ID;".format(key)
        cursor.execute(q)
        cursor.fetchall()
        join_table = pd.read_sql_query(q, connection)
        print('\n', '- ' * 40, '\n')
        print('Data of age has been decrypted:\n')
        print(join_table)
        
except Error as e:
    print('Error while connecting to MySQL', e)

finally:
    if (connection.is_connected()):
        cursor.close()
        connection.close()
        print('\n', '* ' * 40, '\n')
        print('MySQL connection is closed.')
```
